What do you say if we define a range for **Rent** variable so that it will be more easy for customers to search for preferred choice?

```python
def rent_range(data):
    if data.Rent >=1200 and data.Rent <= 10000:
        data['Rent Category'] = 'Affordable'
    elif data.Rent >10000 and data.Rent <= 1000000:
        data['Rent Category'] = 'Medium'
    else:
        data['Rent Category'] = 'Premium'
    return data

data = data.apply(rent_range,axis='columns')
```
