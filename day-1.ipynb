```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = pd.read_csv("House_Rent_Dataset.csv",header=None)
print("the first 5 rows of dataframe")
dt.head(5)
```

```python
dt = dt.iloc[1:]
dt.head(5)
```

```python
print("the bottom 10 rows of dataframe")
dt.tail(10)
```

### Add headers

```python
headers = ["Posted On","BHK","Rent","Size","Floor","Area Type","Area Locality","City","Furnishing Status",
          "Tenant Preferred","Bathroom","Point Of Contact"]
print("headers\n",headers)
```

```python
dt.columns = headers
dt.head(5)
```

Replace "?" with "NaN"

```python
dt1 = dt.replace("?",np.NaN)
```

```python
dt1.head(10)
```

Drop missing values along rent column

```python
# axis = 0 implies along rows and axis = 1 implies along columns
dt1 = dt1.dropna(subset=["Rent"],axis=0)
dt1.head(5)
```
